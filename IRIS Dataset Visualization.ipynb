```python
df.head()
```

```python
df['Species'].unique()
```

```python
plt.hist([df[df.Species == 'Iris-setosa'].SepalLengthCm, \
        df[df.Species == 'Iris-versicolor'].SepalLengthCm, \
        df[df.Species == 'Iris-virginica'].SepalLengthCm], bins=15, stacked=True)
plt.show()
```

```python
list1 = list()
mylabels=list()
for gen in df.Species.cat.categories:
    list1.append(df[df.Species==gen].SepalLengthCm)
    mylabels.append(gen)

hist1 = plt.hist(list1, bins=15, stacked=True, label=mylabels)
plt.legend()
plt.show()
```
